```python
from torch import nn
from torch.nn import functional as F

class GPTForSequenceClassification(nn.Module):
    def __init__(
        self, 
        n_head, 
        block_size, 
        n_embd, 
        n_layer, 
        vocab_size, 
        dropout, 
        device,
        num_labels  # NEW: number of classes
    ):
        super().__init__()
        
        # Original GPT model (feature extractor)
        self.gpt = GPTLanguageModel(
            n_head=n_head,
            block_size=block_size,
            n_embd=n_embd,
            n_layer=n_layer,
            vocab_size=vocab_size,
            dropout=dropout,
            device=device
        )
        
        # Classification head
        self.classifier = nn.Linear(n_embd, num_labels)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, idx, labels=None):
        # Get GPT representations
        # We need to modify GPT to return hidden states
        B, T = idx.shape
        
        tok_emb = self.gpt.token_embedding_table(idx)
        pos_emb = self.gpt.position_embedding_table(
            torch.arange(T, device=self.gpt.device)
        )
        x = tok_emb + pos_emb
        x = self.gpt.blocks(x)
        x = self.gpt.ln_f(x)  # (B, T, n_embd)
        
        # Pool the sequence representation
        # Strategy 1: Use last token (like GPT)
        pooled = x[:, -1, :]  # (B, n_embd)
        
        # Strategy 2: Mean pooling (alternative)
        # pooled = x.mean(dim=1)  # (B, n_embd)
        
        # Strategy 3: Max pooling (alternative)
        # pooled = x.max(dim=1)[0]  # (B, n_embd)
        
        # Apply classification head
        pooled = self.dropout(pooled)
        logits = self.classifier(pooled)  # (B, num_labels)
        
        loss = None
        if labels is not None:
            loss = F.cross_entropy(logits, labels)
        
        return logits, loss
```
